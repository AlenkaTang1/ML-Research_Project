## **Baseline SVM classifier with rbf kernel for kmer = 3, 5, 10**


```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
human_dna = pd.read_table('/content/human.txt')
```

```python
#for Kmer = 3

from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import pandas as pd

#loading the data
df_k3 = pd.read_csv('vector_list_k3.csv',header=0)

human_dna.rename(columns={'class': 'label'}, inplace=True)
y = human_dna.label
X_k3 =df_k3
y = y[:1000]

 # Split the data into training and testing sets (80% for training, 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(X_k3, y, test_size=0.2, random_state=42)
```

```python
# SVM classifier with RBF kernel
classifier_rbf_k3 = SVC(kernel='rbf', random_state=42)
classifier_rbf_k3.fit(X_train, y_train)

#prediction result
prediction_k3 = classifier_rbf_k3.predict(X_test)
accuracy_k3 = accuracy_score(y_test, prediction_k3)
print("accuracy for kmer 3:",accuracy_k3)
```

## **testing rbf kernel with different C and Gamma value**

```python
import numpy as np
import matplotlib.pyplot as plt
#range of parameters
C_range = np.logspace(-2,10,13)
gamma_range = np.logspace(-9,3,13)
```

```python
#for Kmer = 3

from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import pandas as pd

#loading the data
df_k3 = pd.read_csv('vector_list_k3.csv',header=0)

human_dna.rename(columns={'class': 'label'}, inplace=True)
y = human_dna.label
X_k3 =df_k3
y = y[:1000]

 # Split the data into training and testing sets (80% for training, 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(X_k3, y, test_size=0.2, random_state=42)
```

```python
#result storage
accuracy_c_k3 = []
accuracy_gamma_k3 = []

#test over C
for C in C_range:
  classifier_rbf_k3 = SVC(C=C, kernel = 'rbf', gamma = 'scale')
  classifier_rbf_k3.fit(X_train,y_train)
  y_pred = classifier_rbf_k3.predict(X_test)
  accuracy_c_k3.append(accuracy_score(y_test,y_pred))

#test cover gamma
for gamma in gamma_range:
  classifier_rbf_k3 = SVC(C=1, kernel = 'rbf', gamma = gamma)
  classifier_rbf_k3.fit(X_train, y_train)
  y_pred = classifier_rbf_k3.predict(X_test)
  accuracy_gamma_k3.append(accuracy_score(y_test,y_pred))

#plotting
plt.figure(figsize = (15,5))

#plot for C
plt.subplot(1,2,1)
plt.semilogx(C_range,accuracy_c_k3)
plt.title('Accuracy vs C for SVM')
plt.xlabel('C')
plt.ylabel('Accuracy')

#plot for gamma
plt.subplot(1,2,2)
plt.semilogx(gamma_range,accuracy_gamma_k3)
plt.title('Accuracy vs gamma for SVM')
plt.xlabel('gamma')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```
